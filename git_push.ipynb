# 更新到 GitHub

这个笔记本演示如何检查 Git 安装、查看仓库状态，并将本地 `main` 分支推送到远程 `origin`。

```python
import subprocess
from pathlib import Path
repo = Path(r'C:\Users\zdxzh\Desktop\aws-saa-c03-study-notes')
print('Repository path:', repo)
print('Git version:')
print(subprocess.run(['git', '--version'], capture_output=True, text=True).stdout)
print('Current branch and status:')
print(subprocess.run(['git', '-C', str(repo), 'status', '--short', '--branch'], capture_output=True, text=True).stdout)
```

```python
import subprocess
from pathlib import Path
repo = Path(r'C:\Users\zdxzh\Desktop\aws-saa-c03-study-notes')
cmd = ['git', '-C', str(repo), 'push', '-u', 'origin', 'main']
print('Running:', ' '.join(cmd))
proc = subprocess.run(cmd, capture_output=True, text=True)
print('RETURN CODE:', proc.returncode)
print('STDOUT:
', proc.stdout)
print('STDERR:
', proc.stderr)
```
